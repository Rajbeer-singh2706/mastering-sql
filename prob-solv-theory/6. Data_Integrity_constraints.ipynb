## **Q1) What are foreign key constraints, and how do they enforce data integrity?**

### **Foreign Key Constraint:**
A **foreign key** is a column (or a set of columns) in one table that establishes a relationship with the **primary key** of another table.  

‚úÖ **Purpose:**  
- Ensures **referential integrity** by preventing orphan records.  
- Enforces that a referenced value must exist in the parent table before it can be inserted in the child table.  

### **Example: Foreign Key Constraint**
#### **Parent Table (Departments)**
```sql
CREATE TABLE departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(100) NOT NULL
);
```

#### **Child Table (Employees) with a Foreign Key**
```sql
CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(100) NOT NULL,
    dept_id INT,
    FOREIGN KEY (dept_id) REFERENCES departments(dept_id)
);
```
‚úÖ **Enforced Rules:**  
1. You **cannot insert** an `employees` record with a `dept_id` that does not exist in `departments`.  
2. You **cannot delete** a `departments` record if employees exist under it (unless handled with `ON DELETE CASCADE` or `SET NULL`).

## **Q2) Explain the difference between primary keys and unique constraints. Can you have multiple unique constraints on a table?**

### **Primary Key vs. Unique Constraint**

| Feature | Primary Key | Unique Constraint |
|---------|------------|------------------|
| Uniqueness | ‚úÖ Ensures uniqueness | ‚úÖ Ensures uniqueness |
| NULL Allowed? | ‚ùå No (Cannot be NULL) | ‚úÖ Yes (Can have multiple NULL values) |
| Number of Constraints per Table | ‚ùå Only one per table | ‚úÖ Multiple per table |
| Automatically Indexed? | ‚úÖ Yes | ‚úÖ Yes |

‚úÖ **Yes, you can have multiple UNIQUE constraints on a table.**  
- Primary keys enforce **uniqueness & non-nullability**.  
- Unique constraints allow **multiple NULL values** but ensure that non-null values are unique.

### **Example: Primary Key vs. Unique Constraint**
```sql
CREATE TABLE students (
    student_id INT PRIMARY KEY,      -- Only one primary key per table
    email VARCHAR(100) UNIQUE,       -- Multiple unique constraints allowed
    phone_number VARCHAR(15) UNIQUE  -- Another unique constraint
);
```
üîπ **Use PRIMARY KEY for uniquely identifying records.**  
üîπ **Use UNIQUE constraints when allowing NULL values is acceptable.**

## **Q3) What are check constraints, and how would you use them to enforce data validation rules in SQL?**

### **CHECK Constraint:**
‚úÖ **Used to enforce custom rules at the column level.**  
‚úÖ **Ensures that only valid data is inserted.**  

### **Example: CHECK Constraint for Age Validation**
```sql
CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(100) NOT NULL,
    age INT CHECK (age >= 18),   -- Ensures no employee is under 18
    salary DECIMAL(10,2) CHECK (salary > 0)  -- Salary must be greater than 0
);
```

### **Example: CHECK Constraint for Gender Values**
```sql
CREATE TABLE users (
    user_id INT PRIMARY KEY,
    username VARCHAR(100) NOT NULL,
    gender CHAR(1) CHECK (gender IN ('M', 'F', 'O'))  -- Restricts values to M, F, or O
);
```
üîπ **Prevents invalid data entry, reducing the need for additional validation in the application layer.**

## **Q4) How would you handle the deletion of a record that has dependencies via foreign key constraints?**

When deleting a record that is referenced by a foreign key, SQL provides multiple ways to handle it:

| Deletion Strategy      |      Description |
|------------------------|------------|
| **ON DELETE CASCADE**  | Automatically deletes child records when a parent record is deleted. |
| **ON DELETE SET NULL** | Sets foreign key values to NULL instead of deleting child records. |
| **ON DELETE RESTRICT** | Prevents deletion if any dependent records exist. |
| **ON DELETE NO ACTION** | Default behavior (same as RESTRICT). |

---

### **Q5) Explain the difference between ON DELETE CASCADE and ON DELETE RESTRICT.**

‚úÖ **ON DELETE CASCADE:**  
- Automatically deletes child records when the parent is deleted.  
- **Useful when child records should not exist without the parent.**  

#### **Example of CASCADE**
```sql
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    customer_id INT,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id) ON DELETE CASCADE
);
```
```sql
DELETE FROM customers WHERE customer_id = 101;
```
üí° **All orders related to `customer_id = 101` will be deleted automatically.**

---

‚úÖ **ON DELETE RESTRICT:**  
- Prevents deletion if dependent records exist in the child table.  
- **Ensures referential integrity by blocking accidental deletions.**  

#### **Example of RESTRICT**
```sql
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    customer_id INT,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id) ON DELETE RESTRICT
);
```
```sql
DELETE FROM customers WHERE customer_id = 101;
```
‚ùå **Error:** "Cannot delete customer because related records exist in the `orders` table."

‚úÖ **Use ON DELETE CASCADE when you want to remove child records automatically.**  
‚úÖ **Use ON DELETE RESTRICT when you want to prevent accidental deletions.**
